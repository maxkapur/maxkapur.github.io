```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Steve and Dave

Let Steve and Dave be two souls who occupy the same body. This means that they receive identical sensory input, represented here by the function <code>environment</code>. We want to study how Steve and Dave&lsquo;s emotional philosophies help or hurt them in coping with their environment&rsquo;s ups and downs.

<code>environment</code> is just 4000 samples from a normal distribution. Think about this as a composite of all the good and bad things that happen to Steve/Dave. In real life, we are capable of differentiating between different kinds of good and bad, so we recognize a difference between (say) feeling mildly upset in a Prius and crying desparately in a Maserati. But to the extent that these goods and bads balance each other, we can say that quality of our environment from day to day is basically a random function. There might be long-term trends, but our purpose here is too study emotions, not life satisfaction.

The graph below shows thirty consecutive samples of <code>environment</code>. These could represent thirty days, weeks, months, or even years; it doesn't really matter, as the scale would affect only the variance of the normal distribution and not its underlying randomness.

```python
np.random.seed(55555)

environment = np.random.randn(4000)

# Portion to graph
d1 = 50
d2 = 80
t = np.arange(d1,d2)

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111,xlabel="Time",ylabel="")
ax.plot(t,environment[d1:d2],color="black",label="Quality of environment")
ax.legend()
```

While the <code>environment</code> function shown above represents the *objective* quality of Steve/Dave&rsquo;s environment at a given moment in time, Steve and Dave&rsquo;s *subjective* experiences of their environment&mdash;their moods&mdash; also depend on how their environment is changing&mdash;whether things seem to be getting better or worse.

In this regard Steve and Dave possess different emotional philosophies. Steve is a sensitive guy. It doesn&rsquo;t take much to pick him up after a bad day. Dave, for his part, is emotionally disciplined. He&rsquo;s suspicious of quick rushes and derives happiness only from sustained increases in the quality of his environment&mdash;ironic, given that unbeknownst to him, the quality of his environment is determined entirely by chance.

How will Steve and Dave fare in the long term?

In this experiment the environment is randomly determined, so regardless of whatever hopes Steve and Dave may have, they cannot predict how the environment will perform in the future. The best they can do is compare how they are feeling *now* with how they've been feeling *recently.*

We'll perform this comparison the <code>mood</code> function. It takes two inputs: <code>t</code>, the time; and <code>d</code>, the number of recent observations to compare the current environment with.

<code>mood</code> computes the signed difference between the current <code>environment</code> and the average of <code>environment</code> over the previous <code>d</code> observations.

```python
def mood(t,d): return environment[t] - np.mean(environment[t-d:t])

ExampleMoods = np.vectorize(mood)(t,3)

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111,xlabel="Time",ylabel="")
ax.plot(t,environment[d1:d2],color="black",label="environment")
ax.plot(t,ExampleMoods,color="crimson",label="mood",ls="dashed")
ax.legend()
```

Perhaps you can see that <code>mood</code> is just a really shitty estimate for the derivative of <code>environment</code>, one biased highly by the value of <code>environment</code> itself. This reflects the fact that people are really bad at assessing their own prospects.

There's another problem with <code>mood</code>, though. It needs to account for the fact that [humans are more sensitive to losses than gains](https://en.wikipedia.org/wiki/Loss_aversion), by a factor that I'm going to ballpark as <code>3</code>. Let's fix that.

```python
def mood(t,d): 
    a = environment[t] - np.mean(environment[t-d:t])
    if a < 0: return 3 * a
    else: return a

ExampleMoods2 = np.vectorize(mood)(t,3)    
    
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111,xlabel="Time",ylabel="")
ax.plot(t,environment[d1:d2],color="black",label="environment")
ax.plot(t,ExampleMoods,color="crimson",label="mood",ls="dashed")
ax.plot(t,ExampleMoods2,color="crimson",label="mood (with loss aversion)")
ax.legend()
```

OK, now we can get formular about Steve and Dave's feelings.

Both Steve and Dave respond primarily to their <code>environment</code>. If the environmental quality is good, both will generally feel good. But their subjective feelings are also affecting by their <code>mood</code>.

Since Steve is sensitive, a mild positive stimuli yields a generous response in his emotional state. Likewise, a negative interaction that Dave might take in stride can tank Steve's mood for the rest of the day. In other words, Steve doesn't search very widely in assessing his <code>mood</code>. He just compares how he's feeling now with the most previous observation.

Dave, by contrast, is emotionally disciplined. In his <code>mood</code> function, he takes a longer view than Steve by setting the search paramater to <code>d = 10</code>. That is, Dave compares his current mood to the average of his ten most recent moods.

We multiply these adjusted <code>mood</code> values by <code>0.8</code> to make the graph pretty and add the result to the quality of the <code>environment</code> to get Steve and Dave's subjective <code>SteveFeelings</code> and <code>DaveFeelings</code>. Hashtag mood.

```python
def Steve(t): return environment[t] + 0.8 * mood(t,d = 1)
def Dave(t): return environment[t] + 0.8 * mood(t,d = 10)

SteveFeelings = np.vectorize(Steve)(t)
DaveFeelings = np.vectorize(Dave)(t)
    
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111,xlabel="Time",ylabel="")
ax.plot(t,environment[d1:d2],color="black",label="environment")
ax.plot(t,ExampleMoods2,color="crimson",label="mood (with loss aversion)")
ax.plot(t,SteveFeelings,color="seagreen",label="Steve's feelings")
ax.plot(t,DaveFeelings,color="cornflowerblue",label="Dave's feelings")
ax.legend()
```

As you&rsquo;d expect, Steve experiences wilder mood swings. His highs are higher and his lows are lower.

But that&rsquo;s not all. While both Steve and Dave experience loss aversion, Steve's loss aversion is compounded by his overall moodiness, which means that his higher higs *don&rsquo;t* cancels on his lower lows. In fact, as the scatter plot below shows, he ends up feeling a lot worse on average.

(Not that Dave is particularly cheery either. The phenomenon of loss aversion guarantees that we'll always feel bad unless the quality of the environment trends upward over time. This helped a lot with evolution.)

```python
# This time we'll include all 4000 observations.
# You can do even more if you want. It converges on similar averages.

d1 = 50
d2 = 4000

t = np.arange(d1,d2)

SteveFeelings = np.vectorize(Steve)(t)
DaveFeelings = np.vectorize(Dave)(t)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111,xlabel="Environment",ylabel="Steve and Dave’s feelings")
ax.plot(environment[d1:d2],DaveFeelings,color="cornflowerblue",label="Dave’s average: {}".format(np.mean(DaveFeelings)),ls="none",markersize=5,marker="+",alpha=0.6)
ax.axhline(y=np.mean(DaveFeelings),color="cornflowerblue")
ax.plot(environment[d1:d2],SteveFeelings,color="seagreen",label="Steve’s average: {}".format(np.mean(SteveFeelings)),ls="none",markersize=5,marker="+",alpha=0.6)
ax.axhline(y=np.mean(SteveFeelings),color="seagreen")
ax.legend()
```

The upshot is that if you just ask who was feeling better, they are neck and neck. Steve doesn't have *more* bad days than Dave. His bad days are just *worse.*

```python
# Proportion of observations for which Dave is happier than Steve
sum((DaveFeelings - SteveFeelings) > 0) / len(SteveFeelings)
```
