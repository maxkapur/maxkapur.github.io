```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
# Random data
data=pd.DataFrame(np.random.rand(10000,2),columns=['x','y'])
```

```python
# Compute radii
data['r']=np.sqrt((data['x'] ** 2) + (data['y'] ** 2))
# Code for in circle
data['i']=data['r']<1
```

```python
data.head()
```

```python
data['i'].sum()*4/data['i'].count()
```

```python
plt.figure(figsize=(6,6))
plt.axes(xlim=(0,1),ylim=(0,1))
sns.scatterplot('x','y',data=data,hue='i',alpha=.7,marker='+')
plt.legend(loc=(1.05,0.8))
```

```python
# Many trials
s= np.floor(1.1**np.arange(20,150)).astype('i8') # I just like this list of values
results=[]
for i in s:
    data= np.random.rand(i,2)
    a=np.sqrt(data[:,0]**2 + data[:,1]**2)<1
    results.append(sum(a)*4/len(a))
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(s, results)
plt.xscale('log')
plt.axhline(np.pi,color='red',lw=0.5)
plt.xlabel('n')
plt.ylabel('pi est')
```
