```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def yp(y):
    return (y**2 - y - 2)*(1 - y)**2
```

```python
ys= np.linspace(-2,3,51)
```

```python
fig,ax=plt.subplots(figsize=(8,8))
ax.plot(ys,yp(ys))
ax.axhline(0, color='black',lw=1)
ax.axvline(0, color='black',lw=1)
```
